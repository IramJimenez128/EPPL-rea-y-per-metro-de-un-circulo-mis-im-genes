```python
import math
# Calcular area y perimetro de un circulo
# Variables P, A, r
# Formulas A = pi*r^2 y P = 2 * pi * r
# Capturar el valor de r
print("Calcular area y perimetro de un circulo")
r= float(input("Ingrese el valor de radio: "))
A = math.pi*(r**2)
P = 2*math.pi*r
print("El area de el circulo es: ",A)
print("El perimetro de el circulo es: ",P)
print("Iram Jimenez NL = 28 ")
```
